```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model
```

```python
suprafete
```

```python
suprafete.reshape(-1, 1)
```

```python
[ [37, 1994],  [42, 2000],  54,  69,  74, 104]
```

```python
suprafete_reshaped = suprafete.reshape(-1, 1)
```

```python
model.fit(suprafete_reshaped, preturi)
```

```python
model.coef_, model.intercept_
```

```python
a = model.coef_ 
b = model.intercept_
```

```python
suprafete
```

```python
preturi_prezise = a * suprafete + b
preturi_prezise
```

```python
plt.scatter(suprafete, preturi / 1000)
plt.xlabel('suprafete')
plt.ylabel('preturi')
```

```python
plt.scatter(suprafete, preturi)
plt.plot(suprafete, preturi_prezise)
plt.xlabel('suprafete')
plt.ylabel('preturi')
```
