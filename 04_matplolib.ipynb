```python
def f(x):
    a = 1.5
    b = 2 
    return a * x + b  
```

```python
x_array = np.arange(10, 100, 5)
x_array
```

```python
y_calculat_array = np.array([ f(i) for i in x_array ])
y_calculat_array
```

```python
plt.scatter(x, y);
plt.plot(x_array, y_calculat_array);
plt.xlabel(" Coordonata x")
plt.ylabel(" Coordonata y")
```
