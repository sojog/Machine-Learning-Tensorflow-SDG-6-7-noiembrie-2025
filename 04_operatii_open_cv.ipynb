```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
cv2.__version__
```

```python
image_path = "cat.png"
```

```python
cat_image = cv2.imread(image_path)
```

```python
plt.imshow(cat_image)
```

```python
gray_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
plt.imshow(gray_image, cmap="gray")
```

```python
gray_image.shape
```

```python
reshape = ((gray_image.shape[0] // 2,  gray_image.shape[1] // 2 ))
reshape
```

```python
resized_image = cv2.resize(gray_image, reshape )
plt.imshow(resized_image, cmap="gray")
```

```python
blurred_image = cv2.blur(gray_image, (10, 10))
plt.imshow(blurred_image, cmap="gray")
```

```python
blurred_image = cv2.blur(gray_image, (50, 50))
plt.imshow(blurred_image, cmap="gray")
```

```python
blurred_image = cv2.blur(gray_image, (50, 50))
plt.imshow(blurred_image, cmap="gray")
cv2.imwrite("blurred_image.png", blurred_image)
```

```python
top_sobel_kernel = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])
top_sobel_kernel
```

```python
top_sobel_image = cv2.filter2D(gray_image, -1, top_sobel_kernel)
top_sobel_image
```

```python
plt.imshow(top_sobel_image, cmap="gray")
```

```python
left_sobel_kernel = top_sobel_kernel.T
left_sobel_kernel
```

```python
left_sobel_image = cv2.filter2D(gray_image, -1, left_sobel_kernel)
plt.imshow(left_sobel_image, cmap="gray")
```
