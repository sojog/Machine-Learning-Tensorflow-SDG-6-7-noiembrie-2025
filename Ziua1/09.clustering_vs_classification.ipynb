```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np
```

```python
x, y = make_classification(n_samples=30, n_features=2, n_redundant=0,  n_classes=2)
```

```python
x
```

```python
coord_x = x[:, 0]
coord_y = x[:, 1]
```

```python
y
```

```python
plt.scatter(coord_x, coord_y)
```

```python
plt.scatter(coord_x, coord_y, c=y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier()
knn_model
```

```python
knn_model.fit(x, y)
```

```python
plt.scatter(coord_x, coord_y, c=y)
```

```python
knn_model.predict([[-2, -1]])
```

```python
knn_model.predict([[1, -1]])
```
