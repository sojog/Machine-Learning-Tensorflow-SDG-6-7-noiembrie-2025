```python
def functie_activare(input):
    return 1 if input > 0 else 0

def AND_NEURON(x1, x2):
    w1, w2, b = 101, 231, -141 # trebuie gasite
    return functie_activare(x1 * w1 + x2 * w2 + b)

for x1, x2 in ((0, 0), (0, 1), (1, 0), (1, 1)):
    print(f"AND({x1},{x2}) = {AND_NEURON(x1, x2)}")
```

```python
def functie_activare(input):
    return 1 if input > 0 else 0

def AND_NEURON(x1, x2):
    
    w1, w2, b = 1, 1, -1.5 # trebuie gasite

    return functie_activare(x1 * w1 + x2 * w2 + b)

for x1, x2 in ((0, 0), (0, 1), (1, 0), (1, 1)):
    print(f"AND({x1},{x2}) = {AND_NEURON(x1, x2)}")
```

```python
def functie_activare(input):
    return 1 if input > 0 else 0

def OR_NEURON(x1, x2):
    
    w1, w2, b = 1, 1, -0.5 # trebuie gasite

    return functie_activare(x1 * w1 + x2 * w2 + b)

for x1, x2 in ((0, 0), (0, 1), (1, 0), (1, 1)):
    print(f"AND({x1},{x2}) = {OR_NEURON(x1, x2)}")
```
