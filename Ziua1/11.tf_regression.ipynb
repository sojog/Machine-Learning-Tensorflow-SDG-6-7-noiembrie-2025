```python
import tensorflow as tf
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.linspace(0, 10, 5)
```

```python
x = np.linspace(0, 10, 6)
x
```

```python
y = 2 * x + 1
```

```python
plt.scatter(x, y);
```

```python
x_tensor = tf.constant(x)
y_tensor = tf.constant(y)


plt.scatter(x_tensor, y_tensor);
```

```python
tf.keras
```

```python
tf.keras.__version__
```

```python
model_1 = tf.keras.Sequential(
    [
        tf.keras.layers.Dense(1)
    ]
)
model_1
```

```python
model_1.summary()
```

```python
model_1.compile(optimizer='sgd', loss='mse')
```

```python
x
```

```python
x.reshape(-1, 1)
```

```python
x_tensor_reshaped = tf.expand_dims(x_tensor, axis=1)
x_tensor_reshaped
```

```python
model_1.fit(x_tensor_reshaped, y_tensor, epochs=2)
```

```python
plt.scatter(x_tensor, y_tensor);
```

```python
7 * 2 + 1
```

```python
model_1.predict(np.array([[7]]))
```

```python
model_2 = tf.keras.Sequential(
    [
        tf.keras.layers.Dense(1)
    ]
)
model_2.compile(optimizer='sgd', loss='mse')
model_2.fit(x_tensor_reshaped, y_tensor, epochs=10)
```

```python
model_2.predict(np.array([[7]]))
```

```python
model_3 = tf.keras.Sequential(
    [
        tf.keras.layers.Dense(1)
    ]
)
model_3.compile(optimizer='sgd', loss='mse')
model_3.fit(x_tensor_reshaped, y_tensor, epochs=20, verbose=0)
```

```python
model_3.predict(np.array([[7]]))
```

```python
model_4 = tf.keras.Sequential(
    [
        tf.keras.layers.Dense(1)
    ]
)
model_4.compile(optimizer='sgd', loss='mse')
model_4.fit(x_tensor_reshaped, y_tensor, epochs=200, verbose=0)
```

```python
model_4.predict(np.array([[7]]))
```

```python
model_5 = tf.keras.Sequential(
    [
        tf.keras.layers.Dense(1),
         tf.keras.layers.Dense(1)
    ]
)
model_5.compile(optimizer='sgd', loss='mse')
model_5.fit(x_tensor_reshaped, y_tensor, epochs=200, verbose=0)
model_5.predict(np.array([[7]]))
```
